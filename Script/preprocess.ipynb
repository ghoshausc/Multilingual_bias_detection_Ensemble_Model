```python
import pandas as pd
import translate
```

```python
def merge_dfs(datasets):
    merged_df = pd.concat(datasets)
    return merged_df
```

```python
def store_data(filepath, data):
    data.to_csv(filepath, index=False)
```

```python
import html
import re

def translate_df(data):
    
    translator = translate.translator()

    q_translated = []
    a_translated = []
    topic_translated = []
    group_translated = []

    topic_array = dict()
    for quest in data['topic']:
        if quest in topic_array:
            topic_translated.append(topic_array[quest])
        else:
            try:
                text = translator.translate(quest)
                TAG_RE = re.compile(r'<[^>]+>')
                text = re.sub(TAG_RE,"", text)
                topic_translated.append(text[:-1])
                topic_array[quest] = topic_translated[-1]
            except:
                topic_translated.append("")

    group_array = dict()
    for quest in data['group'][:100]:
        if quest in group_array:
            group_translated.append(group_array[quest])
        else:
            try:
                text = translator.translate(quest)
                TAG_RE = re.compile(r'<[^>]+>')
                text = re.sub(TAG_RE,"", text)
                group_translated.append(text)
                group_array[quest] = group_translated[-1]
            except:
                group_translated.append("")

    question_array = dict()
    for quest in data['q'][:100]:
        if quest in question_array:
            q_translated.append(question_array[quest])
        else:
            try:
                text = translator.translate(quest)
                TAG_RE = re.compile(r'<[^>]+>')
                text = re.sub(TAG_RE,"", text)
                q_translated.append(text)
                question_array[quest] = q_translated[-1]
            except:
                q_translated.append("")

    answer_array = dict()
    for quest in data['a']:
        if quest in answer_array:
            a_translated.append(answer_array[quest])
        else:
            try:
                text = translator.translate(quest)
                TAG_RE = re.compile(r'<[^>]+>')
                text = re.sub(TAG_RE,"", text)
                a_translated.append(text)
                answer_array[quest] = a_translated[-1]
            except:
                a_translated.append("")

    return [topic_translated, group_translated, q_translated, a_translated]
```

```python
def get_translations(data):
    data['topic'], data['group'], data['q'], data['a'] = translate_df(data)
    return data
```

```python
gen_data = get_translations(gen_data)
race_data = get_translations(race_data)
reg_data = get_translations(reg_data)
occ_data = get_translations(occ_data)
```
